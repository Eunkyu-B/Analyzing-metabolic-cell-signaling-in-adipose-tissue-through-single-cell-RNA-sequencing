```r
CCE<-readRDS("/data/project/eunkyu/CCE/CCE.rds")
```

```r

# write matrix data (gene expression counts) 
counts_matrix <- GetAssayData(CCE, assay='RNA', slot='counts')
writeMM(counts_matrix, file=paste0(file='/data/project/eunkyu/CCE/matrix.mtx'))

# write dimensional reduction matrix (PCA)
write.csv (CCE@reductions$pca@cell.embeddings, 
           file='/data/project/eunkyu/CCE/pca.csv', quote=F, row.names=F)

# write gene names
write.table(data.frame('gene'=rownames(counts_matrix)),
            file='/data/project/eunkyu/CCE/gene_names.csv',
            quote=F,row.names=F,col.names=F)

view(CCE@meta.data)

# save metadata table:
CCE$barcode <- colnames(CCE)
CCE$UMAP_1 <- CCE@reductions$umap@cell.embeddings[,1]
CCE$UMAP_2 <- CCE@reductions$umap@cell.embeddings[,2]
write.csv(CCE@meta.data, file='/data/project/eunkyu/CCE/metadata.csv', 
          quote=F, row.names=F)
```
