```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scanpy as sc

import loompy as lmp
import anndata
from scipy import io
from scipy.sparse import coo_matrix, csr_matrix
import os
```

```python
X=io.mmread("/data/project/eunkyu/CCE/matrix.mtx")
```

```python
adata= anndata.AnnData(X=X.transpose().tocsr())
```

```python
adata
```

```python
metadata=pd.read_csv("/data/project/eunkyu/CCE/metadata.csv")
```

```python
with open("/data/project/eunkyu/CCE/gene_names.csv",'r') as f:
    gene_names=f.read().splitlines()
```

```python
adata.obs=metadata
adata.obs.index=adata.obs['barcode']
adata.var.index=gene_names
```

```python
adata
```

```python
pca=pd.read_csv("/data/project/eunkyu/CCE/pca.csv")
pca.index=adata.obs.index
```

```python
adata.obsm['X_pca']=pca.to_numpy()
adata.obsm['X_umap']=np.vstack((adata.obs['UMAP_1'].to_numpy(),adata.obs['UMAP_2'].to_numpy())).T
```

```python
adata.obs_names = adata.obs_names.rename('cell')
```
